```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
K = 10
N_range = np.arange(50,1000,30)
beta_range = [0.1,0.2,0.25,0.3]
plot_1={}
for beta in beta_range:
    plot_1[beta] = []
```

```python
for N in N_range:
    for beta in beta_range:
        nps = int(N/K)
        thres = nps/3.0
        failure_prob_shard = 1-binom.cdf(thres, nps, beta)
        no_net_failure_prob = (1-failure_prob_shard)**K
        net_failure_prob = 1 - no_net_failure_prob
        plot_1[beta].append(no_net_failure_prob)
```
